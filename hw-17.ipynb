```python
import numpy
```

```python
i = 0
N = 1000000
prize = 0
while i < N:
    r = numpy.random.randint (0, 1000)
    if r == 777:
        prize += 200
    elif r == 999:
        prize += 100
    elif r == 555:
        prize += 50
    elif r == 333:
        prize += 15
    elif r == 111:
        prize += 10
    elif (r - 77) % 100 == 0:
        prize += 5
    elif (r - 7) % 10 == 0:
        prize += 3
    elif r % 100 == 0:
        prize += 2
    elif r % 10 == 0:
        prize += 1
    i += 1
print(prize / N)
```
